```python
#create a table that lists the average reading score per grade per each school (separate by grade and school)
school_data_complete.head(3)
```

```python
#create a series for each grade using a conditional statement
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]
```

```python
#group each series (column) of math scores by school 
ninth_graders_math_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_math_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_math_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelth_graders_math_scores = twelth_graders.groupby(["school_name"]).mean()["math_score"]
```

```python
#combine the series into a dataframe
#df should list average math scores by grade ({})
Avg_math_scores_by_grade = pd.DataFrame ({
    "9th" : ninth_graders_math_scores,
    "10th" : tenth_graders_math_scores,
    "11th" : eleventh_graders_math_scores,
    "12th" : twelth_graders_math_scores    
})
Avg_math_scores_by_grade.round(2)
```
